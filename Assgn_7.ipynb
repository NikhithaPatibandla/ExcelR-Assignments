```python
#Finding the most sold product

data.groupby("Product")["Quantity"].sum()
```

```python
#Compute average profit margin for each product

data['Profit Margin'] = (data['Profit'] / data['Sales']) * 100
average_profit_margin = data.groupby('Region')['Profit Margin'].mean()
average_profit_margin
```
